```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rc,rcParams
from sklearn.linear_model import LinearRegression
rc('font', weight='bold') 
plt.rcParams['font.size'] = 14
from matplotlib.pyplot import figure
from textwrap import wrap

from google.colab import files

import seaborn as sns
```

```python
uploaded = files.upload() 
Ed_data = pd.read_excel('Education_2019_percentages_compiled_wAvg.xlsx')
Brightwood_data = pd.read_excel('Brightwood_Ed_Data.xlsx')
```

```python
Ed_data.head()
```

```python

categories = ['Exceeding\nExpectations', 'Meeting\nExpectations', 'Partially\nMeeting\nExpectations', 'Not\nMeeting\nExpectations']
N_English = Ed_data.iloc[:,1]
MA_English = Ed_data.iloc[:,2]
N_Math = Ed_data.iloc[:,3]
MA_Math = Ed_data.iloc[:,4]
N_Science = Ed_data.iloc[:,5]
MA_Science = Ed_data.iloc[:,6]

x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars


fig, ax = plt.subplots(figsize=(16,12))
rects1 = ax.bar(x - width/2, N_English*100, width, label='North End',color='k')
rects2 = ax.bar(x + width/2, MA_English*100, width, label='Massachusetts', fill=False, hatch='/')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage of Students', fontweight='bold', fontsize=40)
ax.set_title('English MCAS Performance', fontweight='bold', fontsize=40)
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=25)
plt.yticks(fontsize=25)
ax.legend(prop={'size': 20})

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        rounded = round(height)
        ax.annotate('{}'.format(rounded),
                    xy=(rect.get_x() + rect.get_width() / 2, rounded),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=20)


autolabel(rects1)
autolabel(rects2)


#ax.bar_label(rects1, padding=3)
#ax.bar_label(rects2, padding=3)

fig.tight_layout()
plt.show()
```
