```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rc,rcParams
rc('font', weight='bold') 
plt.rcParams['font.size'] = 14

from google.colab import files

import seaborn as sns
```

```python
uploaded = files.upload() 
data = pd.read_excel('Pynchon Rent Schedules.xlsx')
data.head()
data.columns
```

```python
#Year
t = np.array (data["Unnamed: 0"])

#Monthly Contract Rent Schedules
monthly = np.array(data["Monthly Contract Rent Potential"]) 


#Yearly Contract Rent Schedules
yearly = np.array(data["Yearly Conract Rent Potential"]) 
```

```python
from matplotlib import rc,rcParams
rc('font', weight='bold') 
plt.rcParams['font.size'] = 18

fig,ax = plt.subplots(2,1,figsize=(13,22))
ax[0].plot(t,monthly,'ko-')
ax[0].set_xlabel('Year', fontweight='bold')
ax[0].set_ylabel('Rent Prices ($)', fontweight='bold')
ax[0].set_title('Monthly Pynchon Rent Prices', fontweight='bold')
ax[0].set_xticks(t)

#ax[1].plot(t[3:-1],NEnd_waitlist[3:-1],'bo-')#data available only after 2012
ax[1].plot(t,yearly,'ko-')#data available only after 2012
ax[1].set_xlabel('Year', fontweight='bold')
ax[1].set_ylabel('Rent Prices ($)', fontweight='bold')
ax[1].set_title('Yearly Pynchon Rent Prices', fontweight='bold')
ax[1].set_xticks(t)
```
