```python
#%reset -f
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rc,rcParams
from scipy.stats import pearsonr
rc('font', weight='bold') 
plt.rcParams['font.size'] = 18 #setting fonts of plot labels larger

from google.colab import files
```

```python
uploaded = files.upload() #NEnd All Crimes YDT Percent Change #NEnd Crime Groups YDT Percent Change ... excel files
#North End
NEnd_YTD = pd.read_excel('NEnd All Crimes YDT Value (1).xlsx')
Springfield_YTD = pd.read_excel('Springfield All Crimes YDT Value.xlsx')

```

```python
# columns = NEnd_YTD.loc[:, ~NEnd_YTD.columns.isin(['Category', 'Crime', 'Years'])]
t = np.array(NEnd_YTD["Years"])[:12]
for i in range(len(NEnd_YTD)):
  fig,ax1 = plt.subplots(figsize=(12,7)) 
  row = NEnd_YTD.iloc[i]
  title = NEnd_YTD['Crime']
  new_row = row[4:]
  ax1.plot(t,new_row,'ko-', label = 'North End', marker='o', markersize=15, linewidth=7.0)

  row1 = Springfield_YTD.iloc[i]
  title1 = Springfield_YTD['Crime']
  new_row1 = row1[4:]
  ax2 = ax1.twinx()
  ax2.plot(t,new_row1,'ko--', label = 'Springfield', marker='x', markersize=15, linewidth=4.0)

  ax1.set_xlabel('Year', fontweight='bold')
  ax1.set_ylabel('North End Number of Incidents', fontweight='bold')
  ax1.set_title(title[i], fontweight='bold')

  ax2.set_xlabel('Year', fontweight='bold')
  ax2.set_ylabel('Springfield End Number of Incidents', fontweight='bold')
  ax2.set_title(title[i], fontweight='bold')
  plt.xticks(t) #To ensure each year is seen on the x axis 
  #https://www.delftstack.com/howto/matplotlib/how-to-place-legend-outside-of-the-plot-in-matplotlib/
  fig.legend(loc='upper left', bbox_to_anchor=(0.95,0.95)) 
  plt.tight_layout()
  # fig,ax1 = plt.subplots(figsize=(12,7)) 
  plt.savefig(f"/content/drive/MyDrive/PYTHON_FIGURES/lineplot{i}.png")
plt.clf()
```
